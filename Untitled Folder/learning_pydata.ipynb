```python

## Concatenate multiple files into one data frame

sc_filez = [2015,2016,2017]
sc_colz = ["Year","SeasonName", 
           "RegionName", "DistrictName", "SiteName", "GroupName",
           "GlobalClientId", "Facilitator",
           "TotalCredit", "TotalRepaid_IncludingOverpayments", "LastRepaymentDate"]

data_set_all = []

for yr in sc_filez:
    dtmp = pd.read_csv( "%s\/oaf_sc_%d.csv" % (data_dir, yr)  )
    dtmp["Year"] = yr
    print( dtmp.info() )
    print( dtmp[sc_colz ].head() )
    print( "\n\n")
    data_set_all.append( dtmp[sc_colz ] )

data_set_all = pd.concat( data_set_all, ignore_index=True)
    
print( data_set_all.info() )

data_set_all.groupby(["SeasonName", "RegionName"]).count().plot(kind="barh")


pvt_all = data_set_all.pivot_table(index=["Year", "RegionName"], values=["GlobalClientId"], aggfunc="count")

pvt_all

pvt_all.plot( title="Clients By Region Per year", kind="barh")
```

```python
## Exsmple random walks

import random 
pos = 0
walk = [pos]
axn = [pos]
steps = 1000
for i in np.arange(steps):
    step = 1 if random.randint(0,1) else -1
    pos += step
    walk.append(pos)
    axn.append(step)

## np equiv
ndraws = np.random.randint(0, 2, size=steps) 
nsteps = np.where( ndraws > 0, 1,-1)
nwalks = nsteps.cumsum()

print( "walk native = ", walk.count, 
      "\n walks np = ", nwalks.shape,
      "\n nwalks min = ", nwalks.min(), 
     "\n walks max = ", nwalks.max() )

plt.plot([walk[:100], np.arange(100)])
```
