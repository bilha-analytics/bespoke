```python
### Setup

## 1. Load Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
from dateutil.parser import parse

## stats models
from patsy import dmatrices
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
from sklearn import metrics
from sklearn.cross_validation import cross_val_score

%matplotlib inline 


# TODO: refactor plt.figure irrespective of dist or gmap etc 
def singlePlot_Dist( x ):
    g = sns.kdeplot(x, shade=True)
    plt.figure() 
    
def singlePlot_DistgMap( x ):
    g = sns.FacetGrid(all_data, col="Year", row="RegionName")
    g.map( sns.distplot, x)
    plt.figure() 
    

def load_csv(yr, fprefix, dsdir):
    print( " WORKING on PATH : %s/%s_%d.csv" %(dsdir,fprefix,yr)  )
    dtmp = pd.read_csv( "%s/%s_%d.csv" %(dsdir,fprefix,yr) )
    dtmp["Year"] = yr
    dtmp["xClientID"] = str(yr) + "_"+dtmp.GlobalClientId 
    dtmp["xSiteID"] = str(yr) + "_"+dtmp.SiteName 
    dtmp["xPodID"] = str(yr) + "_"+dtmp.SiteName+"_"+dtmp.SectorName
    dtmp["xGroupID"] = str(yr) + "_"+dtmp.SiteName+"_"+dtmp.GroupName 
    dtmp.index = dtmp.xClientID
    print( "LOADED df " )
    return dtmp[df_cols] 
    

def computed_fields(df):
    df["PctRepaid"] =  np.minimum( 1, (df.TotalRepaid_IncludingOverpayments/df.TotalCredit) )
    df["OverpaymentAmt"] = np.maximum(0, df.TotalRepaid_IncludingOverpayments - df.TotalCredit)
    df["OutstandingAmt"] = np.maximum(0, df.TotalCredit - df.TotalRepaid_IncludingOverpayments )
    df.rename(columns={"TotalCredit":"TXSize", 
                       "TotalRepaid_IncludingOverpayments":"TotalRepaid",
                      }, inplace=True)

## Agg level outcomes = group size, [sum,mean](totalcredit, totalrepaid, outstanding, overpayment), %repaid 
def group_aggz( grplvl, holding_dict ):
    df = all_data.groupby([ grplvl ]).agg({
        "xClientID" : lambda x: np.count_nonzero( np.unique(x)),
        "xGroupID" : lambda x: np.count_nonzero( np.unique(x)), 
        "PctRepaid" : ["mean", lambda x: np.count_nonzero( x < 1 ) ], 
        "TXSize" : ["mean", "sum"],
        "TotalRepaid" : ["mean", "sum"],
        "OutstandingAmt" : ["mean", "sum"],
        "OverpaymentAmt" : ["mean", "sum"],
    })
    
    ### df flatten multi-index cols     
    #colz_0 = df.columns.get_level_values(0).values
    #colz_1 = df.columns.get_level_values(1).values
    #print(' colz 0 = ', colz_0, "\n colz 1 = ", colz_1)

    #print("\n COlums.values = \n", df.columns.values )
    df.columns = ['_'.join(col).strip() for col in df.columns.values]
    
    ### other computed fields
    df.rename(columns={"PctRepaid_<lambda>":"Indiz_NotCompleted", 
                       "xClientID_<lambda>":"xSize",
                       "xGroupID_<lambda>":"NbGroups",
                       "PctRepaid_mean":"PctRepaid",
                       "TXSize_mean":"TXSize",
                       "TotalRepaid_mean":"TotalRepaid",
                       "OutstandingAmt_mean":"OutstandingAmt",
                       "OverpaymentAmt_mean":"OverpaymentAmt",
                      }, inplace=True)
    df["Is_AllCompleted"] = np.where( df.PctRepaid >= 1, 1, 0)
    df["Pct_IndizCompleted"] = 1 - (df.Indiz_NotCompleted/df.xSize)
    
    ### add identifying columns = RegionName, Year
    
    ### update in dict or global variable assign 
    holding_dict[grplvl] = df
    
    return None
    

print("Done")  
```

```python
## 2. Load def fxs
df_name_prefix = "df_sc"
df_cols = ["xClientID", "xSiteID", "xPodID", "xGroupID",
           "Year",
           "RegionName", "DistrictName", 
          "Facilitator", 
          "TotalCredit", "TotalRepaid_IncludingOverpayments", "LastRepaymentDate"]
    
    
## 3. Define key outcomes 
outcomes = ["TXSize", "PctRepaid", "OutstandingAmt", "OverpaymentAmt", ]
outcomes_aggz = ["Is_AllCompleted", "Pct_IndizCompleted"]


## 4. Load dataset 
data_dir = "D:/xAppzor/datasets/oaf"
data_file_prefix = "oaf_sc"
seasons = [2015, 2016, 2017]


all_data = pd.concat(  np.asarray(
    pd.Series( seasons ).apply(
        lambda x: load_csv(x, data_file_prefix, data_dir) ) 
) )



## 5. Computed Fields
computed_fields( all_data )

# datetime strings convert
#all_data["LastRepaymentDate"] = pd.to_datetime(all_data.LastRepaymentDate)

print("\n\n all data = \n" , all_data.info() , 
      "\nIndecies\n", all_data.index.names,
      #"\n\n Head \n", all_data.head(),
     "\n\n Sample", all_data[ all_data.TotalRepaid != all_data.TXSize][ outcomes ].sample(7),
      #"\n\n CFO\n", all_data.groupby(["xSiteID"]).size()
     )



## 6. Aggregates = group size, [sum,mean](totalcredit, totalrepaid, outstanding, overpayment), %repaid
# grouping levels = site, pod, group
grp_levels = ["xSiteID", "xPodID", "xGroupID"] 
aggz_data = {} 
dnul = pd.Series( grp_levels ).apply( lambda x:  group_aggz(x, aggz_data) ) 
```

# Statistical Analysis
## Logit Regression
- Likelihood of an individual defaulting = f(TXSize, GroupSize, AbsAmtOutstanding)

```python
### 1. Individual level

dtmp = all_data[outcomes]
dtmp["Defaulted"] = np.where(dtmp.PctRepaid < 1,1,0)
dtmp.tail()

#prepare 
y, x = dmatrices( 'Defaulted ~ TXSize + OutstandingAmt', dtmp, return_type='dataframe')
print( x.columns,  "\n\n", x.dtypes, "\n\n", y.dtypes  )


#flatten y in 1-D array
y = np.ravel(y)


#logit regress
model = LogisticRegression()
model = model.fit(x,y)

#check accuracy on training set
print("\n accuracy = ", model.score(x,y),
      "\n\n null error rate = ", 
      "\n\n % Defaulted = ", y.mean()
     )

# examine the coefficients
#pd.DataFrame( zip(x.columns, np.transpose(model.coef_) ) ) 

### split data into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=0)
model2 = LogisticRegression()
model2 = model2.fit(x_train, y_train)
#check accuracy on training set
print("\n\n---- TEST DATA ----\n accuracy = ", model2.score(x_train,y_train),
      "\n\n null error rate = ", 
      "\n\n % Defaulted = ", y_train.mean()
     )

### predict class lables for the test set
predicted = model2.predict(x_test)
print("\n\n---- PREDICTED DATA ----\n", predicted,
      "\nx_test = \n", x_test
     )

### generate class probabilities
probz = model2.predict_proba(x_test)
print("\n\n---- CLASS PROBABILITIES ----\n", probz
     )

### confusion matrix and classification report metrics
print("\n\n---- METRICS ----\n", 
      "Confusion Matrx\n", metrics.confusion_matrix(y_test, predicted),
      "\n\nClassification Report\n", metrics.classification_report(y_test, predicted),
     )


### model evaluation via corss-validation 
scorez = cross_val_score( LogisticRegression(), x, y, scoring='accuracy', cv=10)
print("\n\n---- CROSS VALIDATION ----\n", 
      "scores mean = ", scorez.mean,
      "\n\n scores \n", scorez
     )

#histogram
#dtmp.hist()
```

```python

### Predict probability of default for a client with TXSize and outstanding amount
print("\n\n---- PREDIT ----\n", 
      "\n\nClient TXSize = 10000, Outstanding=800 \t p=", model.predict_proba( np.array([1,10000,800]).reshape(1, -1) ),
      "\n\nClient TXSize = 10000, Outstanding=3700\t p=", model.predict_proba( pd.Series([1,10000,3700]).reshape(1, -1) ),      
      "\n\nClient TXSize = 23000, Outstanding=800\t p=", model.predict_proba( pd.Series([1,23000,800]).reshape(1, -1) ),
      "\n\nClient TXSize = 23000, Outstanding=3700\t p=", model.predict_proba( pd.Series([1,23000,3700]).reshape(1, -1) )
     )
```
